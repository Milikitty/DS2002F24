```python
#1 Reading CSV files
import pandas as pd
import numpy as np

acc_players = pd.read_csv("/content/acc_players-2324F.csv")
acc_players.head(5)
```

```python
#2 Basic analysis
acc_players['Unnamed: 13'] = pd.to_numeric(acc_players['Unnamed: 13'], errors='coerce')
acc_players['Unnamed: 13'] = acc_players['Unnamed: 13'].fillna(0)
acc_players['Unnamed: 13'] = acc_players['Unnamed: 13'].astype(int)
total_points = acc_players['Unnamed: 13'].sum()
print(total_points)
```

```python
#2 Basic analysis
acc_players['Unnamed: 6'] = pd.to_numeric(acc_players['Unnamed: 6'], errors='coerce')
acc_players['Unnamed: 6'] = acc_players['Unnamed: 7'].fillna(0)
acc_players['Unnamed: 6'] = acc_players['Unnamed: 7'].astype(int)
top_minutes_player = acc_players.nlargest(1, 'Unnamed: 6')['Unnamed: 1']
print(top_minutes_player)
```

```python
#2 Basic analysis
acc_players['Unnamed: 7'] = pd.to_numeric(acc_players['Unnamed: 7'], errors='coerce')
acc_players['Unnamed: 7'] = acc_players['Unnamed: 7'].fillna(0)
acc_players['Unnamed: 7'] = acc_players['Unnamed: 7'].astype(int)
top_5_rebounders = acc_players.nlargest(5, 'Unnamed: 7')['Unnamed: 1']
print(top_5_rebounders)
```

```python
#3 Player filtering
acc_players['Unnamed: 6'] = pd.to_numeric(acc_players['Unnamed: 6'], errors='coerce')
acc_players['Unnamed: 6'] = acc_players['Unnamed: 6'].fillna(0)
acc_players['Unnamed: 6'] = acc_players['Unnamed: 6'].astype(int)
filtered_players = acc_players.groupby('Unnamed: 1').agg(
    Filtered_Values=('Unnamed: 6', lambda x: list(x) if sum(x) > 500 else None)
).reset_index()

filtered_players = filtered_players.dropna(subset=['Filtered_Values'])
print(filtered_players)
```

```python
acc_players['Unnamed: 8'] = pd.to_numeric(acc_players['Unnamed: 8'], errors='coerce')
acc_players['Unnamed: 8'] = acc_players['Unnamed: 8'].fillna(0)
acc_players['Unnamed: 8'] = acc_players['Unnamed: 8'].astype(int)

top_assists = acc_players.groupby('Unnamed: 1').agg({'Unnamed: 8': 'sum'})
top_assists = top_assists.nlargest(3, 'Unnamed: 8')
print(top_assists)
```

```python
acc_players['Unnamed: 10'] = pd.to_numeric(acc_players['Unnamed: 10'], errors='coerce')
acc_players['Unnamed: 10'] = acc_players['Unnamed: 10'].fillna(0)
acc_players['Unnamed: 10'] = acc_players['Unnamed: 10'].astype(int)

top_block = acc_players.groupby('Unnamed: 1').agg({'Unnamed: 10': 'sum'})
top_block = top_block.nlargest(3, 'Unnamed: 10')
print(top_block)
```

```python
#4 School Analysis
acc_players['Unnamed: 13'] = pd.to_numeric(acc_players['Unnamed: 13'], errors='coerce')
acc_players['Unnamed: 13'] = acc_players['Unnamed: 13'].fillna(0)
acc_players['Unnamed: 13'] = acc_players['Unnamed: 13'].astype(int)

acc_players['Unnamed: 8'] = pd.to_numeric(acc_players['Unnamed: 8'], errors='coerce')
acc_players['Unnamed: 8'] = acc_players['Unnamed: 8'].fillna(0)
acc_players['Unnamed: 8'] = acc_players['Unnamed: 8'].astype(int)

school_stats = acc_players.groupby('Unnamed: 4').agg({'Unnamed: 13': 'sum', 'Unnamed: 8': 'sum'})
school_stats
```

```python
#4
school_rank = school_stats.sort_values(by='Unnamed: 13', ascending=False)
school_rank.head(3)
```
